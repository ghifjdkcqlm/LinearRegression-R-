lets make a new data frame 

```python
again = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
again
```

a function to be able to remove NaN 

```python
again.dropna()
```

however this dropped all the rows which had at least 1 na 

```python
again.dropna(how = 'all' )
```

above function drops a row only if all the entries of the row are na 

suppose we make one full row na and then wish to eliminate the entire row 

```python
again.iloc[1, again.columns.get_loc('born')] = 'NaN'
```

```python
again
```

now we will drop one entire column from our data frame

```python
again.dropna(how='all',axis = 1)
```

```python
again
```

to remove duplicates from your data frame 

```python
again.drop_duplicates(subset = ['born'])
```

the rows where duplicate values stood in the column of born have been removed
