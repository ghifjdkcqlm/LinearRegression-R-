we can style the lines and the markers like color, linestyle marksize etc

```python
plt.plot(years, apples, marker = 'o', c='r', ls ='-',lw=3,alpha = 0.5, ms = 8)
plt.plot(years, grapes, marker = '*',c = 'g', ls = '--',lw=3, alpha = 0.5, ms = 8)
plt.xlabel('years')
plt.ylabel('yield per hectare')
plt.title('crop yields per year in kanto')
plt.legend(['apples','grapes'])
```

Scatter plot 
values of 2 variables are plotted as points on a 2d grid 

let's take one classic example of  iris flower which is an inbuilt seaborn model 

```python
flowers_df = sns.load_dataset('iris')
```

```python
flowers_df
```

is there a relation between sepal_length and sepal_width  
we will see by plotting a graph between flower_df.sepal_length and flowers_df.sepal_width 

```python
plt.plot(flowers_df.sepal_length ,flowers_df.sepal_width )
```

the above plot dosent look much helpfull so we can use a scatter plot to visualilse how sepal length and sepal width vary using the scatterplot fn from seaborn library  imported as sns 
in scatter plot we will have points instead of these lines 

```python
import seaborn as sns
```

```python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
```

if you pay enough attention youll be able to see three clusters in here which makes sense as there are three species of flowers so similar species showing similar sepal length and width is not very surprising 

so now suppose we want color coding for sepal length and widht plot for each specie so in that case we will add hue 

```python
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width,
                hue = flowers_df.species
                
                );
```

so our theory was correct. similar specie flower due form a cluster here

```python
plt.figure(figsize=(12,6))
plt.title('sepal Dimensions')
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width,
                hue = flowers_df.species
                
                );
```

importing data from pandas data frame for plotting via seaborn

BAR PLOTs with average 

to visualize how the average bill amount varies across different days of the week we can use the barplot fn provided by seaborn library which can automatically compute averages 

```python
tips_df = sns.load_dataset("tips");
```

```python
tips_df
```

how does the average bill look like on tue wed thu ? or any particular time of the week 

```python
bill_avg_df = tips_df.groupby('day')[['total_bill']].mean()
```

```python
bill_avg_df
```

above is the average total bill for thursday, friday, saturday and sunday 

```python
plt.bar(bill_avg_df.index, bill_avg_df.total_bill)
```

one method was the above one, the other way is to directly use the barplot function 


```python
sns.barplot(x = 'day',y = 'total_bill',data=tips_df)
```

```python
sns.barplot(x = 'day',y = 'total_bill',hue = 'sex',data=tips_df)
```

```python
sns.barplot(x = 'day',y = 'total_bill',hue = 'smoker',data=tips_df)
```

to make thee bar horizontal you just have to switch the axes 

```python
sns.barplot(x = 'total_bill',y = 'day',hue = 'smoker',data=tips_df)
```

HEAT MAP 

used for visualising 2d data like a matrix or a table using colors 
we'll use another sample dataset from seaborn called flights to visualise monthly passenger footlball at an airport over 12 years. 

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

presented data set shows footfall at specified months of years in some airport 
suppose we want to know if the footfall increased or decreased over the years we can plot a line chart or a bar chart or a heat map 

```python
plt.plot(flights.year,flights.passengers)
```

overall the trend has been increasing
to get a better visualisation of this we can use a heatmap which tries and puts our data into a matrix kind of a format 

```python
flights = sns.load_dataset('flights')
```

```python
# Load the flights dataset
flights = sns.load_dataset("flights")

```

we use the.pivot_table() fn to convert our data set into a matrix

```python
flights = flights.pivot_table(index="month", columns="year", values="passengers")
```

```python
flights
```

to use a heatmap step one is to bring the dataset into above format and then apply heatmap 

```python
plt.title('No.of passengers (1000s)')
sns.heatmap(flights)
```

brighter colors indicate greater footfall at the airport. 
we can change the color pallette with cmap = 'Blues'  and dispolay original values in each block by specifying annot = True

```python
plt.title('No.of passengers (1000s)')
sns.heatmap(flights, annot = True, cmap = 'Blues')

plt.figure(figsize=(12,12))
```
