now here in this array as you can see there are some non numeric values that are going to cause probelm for us so what we're going to do is we will replace those with null values 
we will use the pd.to_numeric() fn which takes a series/column  and converts the series into numeric data and wherever it encounters a string it will show error but what we want is not the error but a nan value whenever a string is encountered 

```python
survey_df['Age1stCode']= pd.to_numeric(survey_df.Age1stCode, errors = 'coerce')
```

error='coerce': The error='coerce' parameter is used to handle errors during conversion. When error is set to 'coerce', it means that if any value in the column cannot be converted to a numeric type, Pandas will replace that value with NaN (Not a Number).


This code block utilizes Pandas' pd.to_numeric() function to convert the values in the 'Age1stCode' column of the DataFrame survey_df into numeric format

to refine our data we are going to nullify all the multiple entries. that is, only those entries where none or one gender was picked will be considered. We are going to replace all those will empty values 

import numpy as np

```python
survey_df.where(~(survey_df.Gender.str.contains(';',na=False)), np.nan, inplace = True)
```

h.where(): This function is used to replace values in the DataFrame based on a condition. It keeps the original values where the condition is True and replaces the values where the condition is False.

survey_df.Gender.str.contains(';', na=False): This part checks whether each value in the 'Gender' column contains a semicolon (';'). It returns True for rows where the 'Gender' column contains a semicolon and False otherwise.

~: The tilde (~) symbol is a logical NOT operator, which negates the result of the condition. So, it turns True to False and vice versa.
